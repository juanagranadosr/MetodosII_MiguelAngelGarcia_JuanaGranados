Considerando la ecuaciòn de onda:

$$
\frac{\partial^2u}{\partial t^2} = \alpha^2 \frac{\partial^2u}{\partial x^2}
$$

Discretizando:


$$
\frac{u_{i}^{l+1}-2u_{i}^{l}+u_{i}^{l-1}}{(\Delta t)^2} = \alpha ^2 (\frac{u_{i+1}^{l}-2u_{i}^{l}+u_{i-1}^{l}}{(\Delta x)^2})
$$

$$
u_{i}^{l+1}-2u_{i}^{l}+u_{i}^{l-1}=(\Delta t)^2\alpha ^2 (\frac{u_{i+1}^{l}-2u_{i}^{l}+u_{i-1}^{l}}{(\Delta x)^2}) 
$$



Tomando $\lambda = \frac{\alpha \Delta t}{\Delta \rho }$:

$$
\epsilon_{i}^{l+1}-2\epsilon_{i}^{l}+\epsilon_{i}^{l-1}=\lambda ^2 (\epsilon_{i+1}^{l}-2\epsilon_{i}^{l}+\epsilon_{i-1}^{l})
$$

Usando el hecho de que las componentes de Fourier varian en el espacio como:
$$
u_{i\pm 1}^{l}= e^{\pm jk\Delta x}u_{i}^{l}
$$
Y tambièn teniendo en cuenta que las componentes de Fourier varìan en el tiempo como:
$$
u_{i}^{l\pm 1}= e^{\pm jw\Delta t}u_{i}^{l}
$$

Ademàs, para los epsilon se tiene que $\epsilon_{i}^{l}=e^{jw\Delta t}e^{ jk\Delta x}=\psi^l e^{ jk\Delta x}$
Por lo tanto, la ecuaciòn queda de la siguiente manera:

$$
\psi^{l+1} e^{ jk(i)\Delta x}-2\psi^{l} e^{ jk(i)\Delta x}+\psi^{l-1} e^{ jk(i)\Delta x}=\lambda ^2 (\psi^{l} e^{ jk(i+1)\Delta x}-2\psi^{l} e^{ jk(i)\Delta x}+\psi^{l} e^{ jk(i-1)\Delta x})
$$





Dividiendo entre $\psi^{l-1}$:
$$
\psi^2 e^{ jk(i)\Delta x}-2 e^{ jk(i)\Delta x}\psi+ e^{ jk(i)\Delta x}=\lambda ^2 (e^{ jk\Delta x}-2 + e^{- jk\Delta x})e^{ jk(i)\Delta x}\psi
$$

Dividiendo entre $e^{ jk(i)\Delta x}$, la ecuaciòn queda de la siguiente manera:


$$
\psi^2-2\psi +1  =\lambda ^2 ( e^{ jk\Delta x}-2 + e^{ -jk\Delta x})\psi
$$





Usando la identidad de euler $e^{ \pm jnz}=\cos(nz) \pm \sin(nz)$

$$
\psi^2-2\psi +1 =\lambda ^2 (\cos(k\Delta x) + \sin(k\Delta x)-2+\cos(k\Delta x)- \sin(k\Delta x))\psi
$$
$$
\psi^2-2\psi +1 =\lambda ^2 (2\cos(k\Delta x) -2) \psi
$$

$$
\psi^2-2\psi +1 =\lambda ^2 (-4\sin^2(\frac{k\Delta x}{2}))\psi
$$

Igualando a cero


$$
\psi^2+(\lambda ^2 4\sin^2(\frac{k\Delta x}{2})-2)\psi +1 =0
$$

Usando la ecuaciòn cuadratica:

$$
\psi = \frac{{-(\lambda^2 \sin^2(\frac{k\Delta x}{2}) - 2) \pm \sqrt{{(\lambda^2 4\sin^2(\frac{k\Delta x}{2}) - 2)^2 - 4 \cdot 1 \cdot 1}}}}{{2}}
$$

Sin embargo, hay una condiciòn que debe cumplir para que $\psi$ sea un nùmero complejo ya  que se requiere que el error decrezca:

$$
{(\lambda^2 4\sin^2(\frac{k\Delta x}{2}) - 2)^2 - 4 \cdot 1 \cdot 1} <0
$$

$$
(\lambda^2 4\sin^2(\frac{k\Delta x}{2}) - 2)^2 <4
$$

Elevando a ambos lados a la $\frac{1}{2}$ y posteriormente sumando $2$:
$$
\lambda^2 4\sin^2(\frac{k\Delta x}{2})<4
$$
Dado que el rango de $\sin^2(\frac{k\Delta x}{2})$ va de $[-1,1]$, se toma el valor màximo para verificar que se sigue cumpliendo la desigualdad.

$$
\lambda^2 \cdot 4<4
$$

Dividiendo a ambos lados entre 4:

$$
\lambda^2 <1
$$

Elevando a la un medio a ambos miembros, nos da la definiciòn de valor absoluto:

$$
\left|\lambda\right| <1
$$
