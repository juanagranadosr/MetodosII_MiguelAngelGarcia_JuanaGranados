```python
%matplotlib notebook
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import axes3d
import matplotlib as mpl
import matplotlib.animation as animation
from tqdm import tqdm
```

```python
Min, Max, N = 0, 0.2, 201

x = np.linspace(Min,Max,N)
y = x.copy()
h =  x[1] - x[0]
print(h)

def h1(y):
    return 10.

def h2(y):
    return 0.

def h3(x):
    return 0.

def h4(x):
    return 0.

def InitT():
    
    T = np.zeros((N,N))
        
    T[0,:] = h1(y)
    T[-1,:] = h2(y)
    T[:,0] = h3(x)
    T[:,-1] = h4(x)
        
    return T
```

#### Condiciones de Frontera


```python
T = InitT()
#print(T[0,:,:])
T
```

```python
def Relajacion(T,Nit = int(1e4),omega=0.5,tolerancia = 1e-3):
    
    itmax = 0
    
    for it in tqdm(range(Nit)):
        
        dmax = 0.
        
        for i in range(1, len(x)-1):
            for j in range(1, len(y)-1):
                
                #tmp = (T[i, j+1]-2*T[i,j]+T[i,j-1])/h**2 + (T[i+1, j]-2*T[i,j]+T[i-1,j])/h**2
                tmp = (T[i,j+1]+T[i,j-1]+T[i-1,j]+T[i+1,j])/4 - np.exp(x[i]*y[j])*h**2
                
                
                # Matriz resta
                r = omega*(tmp - T[i,j])
                
                # Actualizo la matrix
                T[i,j] += r
                
                if np.abs(r) > dmax:
                    dmax = r
                    
        #print(np.round(T,2))
                
        
        if np.abs(dmax) < tolerancia:
            
            print(it)
            itmax = it
            break
            
    return T,itmax
```

```python
T = InitT()
Tf1,_ =  Relajacion(T)

Tf1
```

```python
fig = plt.figure(figsize=(10,5))
ax = fig.add_subplot(121)
ax1 = fig.add_subplot(122, projection='3d')

X,Y = np.meshgrid(x,y)

c = ax.contourf(X,Y,Tf1.T)

ax1.plot_surface(X,Y,Tf1.T)

plt.xlabel('[m²]')
plt.ylabel('[m²]')

plt.show()
```
