Considere la siguiente ecuación diferencial:
$$\frac{du}{dt}=\alpha u, u(0)=u_0$$
Pruebe que aplicando iterativamente se obtiene 
$$u_k=(1+\alpha\Delta t)^k u_0$$
para el caso de $\alpha<0$.

Para realizar esta demostración se hará el uso del método de inducción en $k$.

- Caso Base. $k=0, u_0=(1+\alpha\Delta t)^0 u_0 = u_0 = u(0).$
- Hipótesis de Inducción. Suponga que $u_n =(1+\alpha\Delta t)^n u_0,\alpha<0$ se satisface.
- Paso Inductivo. Para desarrollar el paso inductivo se utilizará el método de Euler, el cual establece una relación recursiva de la forma $u_{n+1}=u_n+u_n\alpha\Delta t$ tomando $h=\Delta t$ y $u_n' = u_n\alpha$. Considere entonces el paso de $k=n+1$ $$u_{n+1} = u_n(1+\alpha\Delta t)$$ por la hipótesis de inducción $$u_{n+1} = (1+\alpha\Delta t)^n u_0(1+\alpha\Delta t)=(1+\alpha\Delta t)^{n+1} u_0$$

Mostrando así que se cumple para todo $n$.

Ahora bien, intuitivamente, la solución oscillará si $\Delta t>\frac{-1}{\alpha}$, pues de esta desigualdad se tiene que $$1+\alpha\Delta t<0$$ si tomamos $$(1+\alpha\Delta t)^k<0$$ esta desigualdad se cumple únicamente para el caso de $k=2n+1$, pues si $k$ es par, las soluciones serán positivas. Por lo que se sigue un comportamiento oscilatorio.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def Euler(u0, alfa, delta_t, k):
    
    t = np.arange(0,k,delta_t)
    u = np.zeros_like(t)
    
    for i in range(len(t)):
        u[i]= u0*(1+alfa*delta_t)**i
        
    return u, t
```

```python
def Exacta(alfa, T, u0 ):
    return u0*np.exp(alfa* T)
```

```python
k = 63.
alfa = -1.
u0 = 469.
t = np.linspace(0, k, 100)
u1, t1 = Euler(u0, alfa, 1.1, k)
u2, t2 = Euler(u0, alfa, 1.5, k)
u3, t3 = Euler(u0, alfa, 1.9, k)
UE = Exacta(alfa, t, u0)
```

```python
plt.plot(t1, u1, label='$\Delta T = 1.1$')
plt.plot(t2, u2, label ='$\Delta T = 1.5$') 
plt.plot(t3, u3, label ='$\Delta T = 1.9$')
plt.plot(t, UE, color='red', label='Solución Exacta')
plt.legend()
```
