
Figura 2.4: Diagrama de posiciones Nave-Luna

$c)$ Muestre usando la Figura 2.4 que la distancia Nave-Luna está dada por:

La distancia de la Tierra a la Luna tiene una magnitud igual a:
$$ d = \sqrt{x_{L}^2+y_{L}^2}$$
La vector de posición está definido como:
$$ D_{T-L} = (d\cos{wt},d\sin{wt})$$

La distancia de la Tierra a la Nave tiene una magnitud igual a:
$$ r(t) = \sqrt{x(t)^2+y(t)^2}$$
La vector de posición está definido como:
$$ D_{T-N} = (r(t)\cos{\phi},r(t)\sin{\phi})$$

Restando el vector de posición de la Luna con el vector de posición de la nave, da como resultado:

$$D_{N-L}(r, \phi, t) = (d\cos{wt}-r(t)\cos{\phi},d\sin{wt}-r(t)\sin{\phi})$$

Encontrando la mangnitud, es decir, la distancia entre la Nave y la Luna:

$$r_{N-L}(r, \phi, t) = \sqrt{(d\cos{wt}-r(t)\cos{\phi})^2+d\sin{wt}-r(t)\sin{\phi})^2}$$

Exapndiendo ambos polinomios:

$$r_{N-L}(r, \phi, t) = \sqrt{d^2\cos^2{wt}-2r(t)d\cos{\phi}\cos{wt}+r^2(t)\cos^2{\phi}+d^2\sin^2{wt}-2r(t)d\sin{\phi}\sin{wt}+r^2(t)\sin^2{\phi}}$$
Simplicando usando $rsin^2(\theta)+rcos^2(\theta)=r$ y factorizando:

$$r_{N-L}(r, \phi, t) = \sqrt{d^2+r^2(t)-2r(t)d(\cos{\phi}\cos{wt}+\sin{\phi}\sin{wt})}$$

Dado que $\cos{\phi}\cos{wt}+\sin{\phi}\sin{wt}=\cos{\phi-wt}$. Esto da como resultado:

$$r_{N-L}=(r, \phi, t) = \sqrt{d^2+r^2(t)-2r(t)d(\cos{\phi-wt})}$$


$d)$ Usando esta distancia muestre que el Hamiltoniano de la nave está dado por:

\begin{equation*}
    H = p_r \dot{r} + p_{\phi} \dot{\phi} - L = \frac{p_r^2}{2m} + \frac{p_{\phi}^2}{2mr^2} - G \frac{m m_T}{r} - G \frac{m m_L}{r_L(r,\phi,t)}
\end{equation*}

Dado que el Lagrangiano se define como $L=T-U$, donde $T$ es la energía cinética y $U$ la energía potencial, entonces para hallar $L$, es necesario encontrar la energía cinética y potencial.

Teniendo en cuenta que la posición del cohete respecto a la Tierra (origen) es:
$$\vec{r} = (r(t)\cos{\phi},r(t)\sin{\phi})$$

Su derivada estaría dada por:

$$\vec{\dot{r}} = (\dot{r(t)}\cos{\phi}-r(t)\sin{\phi}\dot{\phi},\dot{r(t)}\sin{\phi}+r(t)\cos{\phi}\dot{\phi})$$

$$\dot{r}^2 = (\dot{r(t)}\cos{\phi}-r(t)\sin{\phi}\dot{\phi})^2+(\dot{r(t)}\sin{\phi}+r(t)\cos{\phi}\dot{\phi})^2$$

$$\dot{r}^2 = \dot{r(t)}^2\cos^2{\phi}-2\dot{r(t)}\cos{\phi}r(t)\sin{\phi}\dot{\phi}+r(t)^2\sin^2{\phi}\dot{\phi}^2+\dot{r(t)}^2\sin^2{\phi}+ 2\dot{r(t)}\cos{\phi}r(t)\sin{\phi}\dot{\phi}+r(t)^2\cos^2{\phi}\dot{\phi}^2$$

$$\dot{r}^2 =\dot{r(t)}^2 + r(t)^2\dot{\phi}^2$$

Dado que $\dot{r}^2=v^2$, la energía cinetica de la nave queda de la siguiente manera:

$$T=\frac{1}{2}m\dot{r}^2=\frac{1}{2}m(\dot{r(t)}^2 + r(t)^2\dot{\phi}^2)$$

Ahora, la energía potencial de la nave en el campo gravitatorio de la Tierra y de la Luna esta dado por la derivada de cada coordenada generalizada, es decir, :

$$U = -G\frac{mM_{T}}{r} -\frac{mM_{l}}{r_{L}}$$

Por lo tanto, el Lagrangiano queda como:

$$L = \frac{1}{2}m(\dot{r(t)}^2 + r(t)^2\dot{\phi}^2)+G\frac{mM_{T}}{r} +\frac{mM_{L}}{r_{L}}$$

Ahora para encontrar el Hamiltoniano, es necesario derivar el Lagrangiano por las derivadas de las coordenadas generalizadas y luego multiplicar esta derivada por la derivada de cada coordenada generalizada, es decir, los momento conjugados.

$$\frac{\partial L}{\partial \dot{\phi}}=p_{r}=mr^2\dot{\phi}$$

$$\frac{\partial L}{\partial \dot{r}}=p_{r}=m\dot{r}$$

Dado que el Hamiltoniano esta definido como:

$$H = \sum_i \frac{\partial L}{\partial \dot{q_i}}\dot{q_i}-L$$

$$H=mr^2\dot{\phi}\dot{\phi}+m\dot{r}\dot{r}-(\frac{1}{2}m(\dot{r(t)}^2 + r(t)^2\dot{\phi}^2)+G\frac{mM_{T}}{r} +\frac{mM_{L}}{r_{L}})$$

$$ H=mr^2\dot{\phi}^2+m\dot{r}^2-(\frac{1}{2}m(\dot{r(t)}^2 + r(t)^2\dot{\phi}^2)+G\frac{mM_{T}}{r} +\frac{mM_{L}}{r_{L}}) $$

$$ H=\frac{1}{2}m(\dot{r(t)}^2 + r(t)^2\dot{\phi}^2)-G\frac{mM_{T}}{r} -\frac{mM_{L}}{r_{L}}) $$

$$H =  \frac{p_r^2}{2m} + \frac{p_{\phi}^2}{2mr^2} - G \frac{m m_T}{r} - G \frac{m m_L}{r_L(r,\phi,t)}$$

$e)$ Muestre que las ecuaciones de Hamilton, que son las ecuaciones de movimiento están dadas por:

Tomando el Hamiltoniano encontrado en el punto anterior:

$$\dot{r}=\frac{\partial (\frac{p_r^2}{2m} + \frac{p_{\phi}^2}{2mr^2} - G \frac{m m_T}{r} - G \frac{m m_L}{r_L(r,\phi,t)})}{\partial p_r}=\frac{p_r}{m}$$

$$\dot{\phi}=\frac{\partial (\frac{p_r^2}{2m} + \frac{p_{\phi}^2}{2mr^2} - G \frac{m m_T}{r} - G \frac{m m_L}{r_L(r,\phi,t)})}{\partial p_{\phi}}=\frac{p_\phi}{mr^2}$$

$$\dot{p_r}=-\frac{\partial (\frac{p_r^2}{2m} + \frac{p_{\phi}^2}{2mr^2} - G \frac{m m_T}{r} - G \frac{m m_L}{r_L(r,\phi,t)})}{\partial {r}}$$

Dado que $r_L(r, \phi, t) = \sqrt{d^2+r^2(t)-2r(t)d(\cos{\phi-wt})} $
$$
\dot{p_r} = \frac{p_{\phi}^2}{mr^3} - \frac{GmM_T}{r^2} + \frac{Gmm_L}{\frac{-1}{2}[\sqrt{d^2 + r^2(t) - 2r(t)d\cos{(\phi - wt)}}]^3}(2r-2d\cos{(\phi-wt)})
$$
$$
\dot{p_r} = \frac{p_{\phi}^2}{mr^3} - \frac{GmM_T}{r^2} - \frac{Gmm_L}{[\sqrt{d^2 + r^2(t) - 2r(t)d\cos{(\phi - wt)}}]^3}(r-d\cos{(\phi-wt)})
$$
$$
\dot{p_r} = \frac{p_{\phi}^2}{mr^3} - \frac{GmM_T}{r^2} - \frac{Gmm_L}{[r_L(r, \phi, t)]^3}(r-d\cos{(\phi-wt)})
$$





$$\dot{p_\phi}=-\frac{\partial (\frac{p_r^2}{2m} + \frac{p_{\phi}^2}{2mr^2} - G \frac{m m_T}{r} - G \frac{m m_L}{r_L(r,\phi,t)})}{\partial {\phi}}$$

$$\dot{p_\phi}=-\frac{\partial (\frac{p_r^2}{2m} + \frac{p_{\phi}^2}{2mr^2} - G \frac{m m_T}{r} - G \frac{m m_L}{\sqrt{d^2+r^2(t)-2r(t)d(\cos{\phi-wt})}})}{\partial {\phi}}$$

$$\dot{p_\phi}=\frac{-1}{2} G \frac{m m_L}{[\sqrt{d^2+r^2(t)-2r(t)d(\cos{\phi-wt})}]^3}(-(2r(t)d[-\sin{(\phi-wt)}]))$$


$$\dot{p_\phi}= -G \frac{m m_L}{[\sqrt{d^2+r^2(t)-2r(t)d(\cos{\phi-wt})}]^3}(r(t)d[\sin{(\phi-wt)}])$$


$$\dot{p_\phi}= -G \frac{m m_L}{[r_L(r, \phi, t)]^3}(r(t)d[\sin{(\phi-wt)}])$$

$f)$ Para reducir el error de redondeo se pueden definir nuevas variables normalizadas a la distancia lunar: $\tilde{r} = r/d$, $\phi$, $\tilde{p_r} = p_r/md$ y $\tilde{p_{\phi}} = p_{\phi}/md^2$. Muestre que el sistema se puede escribir como sigue:


Para $\dot{r}=\frac{p_r}{m}$, se divide a ambos lados entre $d$:


$$\frac{\dot{r}}{d}=\frac{p_r}{dm}$$

Esto es igual a:

$$\dot{\tilde{r}}=\tilde{p_r}$$

Para $\dot{\phi}=\frac{p_\phi}{mr^2}$. Se despeja $p_{\phi}$ de $\tilde{p_{\phi}} = p_{\phi}/md^2$

$$p_{\phi}=\tilde{p_{\phi}}md^2$$

Reemplazando en $\dot{\phi}=\frac{p_\phi}{mr^2}$:

$$\dot{\phi}=\frac{\tilde{p_{\phi}}d^2}{r^2}$$

Dado que $\frac{d^2}{r^2}=\frac{1}{\tilde {r}^2}$

$$\dot{\phi}=\frac{\tilde{p_{\phi}}}{\tilde {r}^2}$$

Para $
\dot{p_r} = \frac{p_{\phi}^2}{mr^3} - \frac{GmM_T}{r^2} - \frac{Gmm_L}{[r_L(r, \phi, t)]^3}(r-d\cos{(\phi-wt)})
$

$$
\dot{p_r} = \frac{p_{\phi}^2}{mr^3} - \frac{GmM_T}{r^2} - \frac{Gmm_L}{[\sqrt{d^2+r^2(t)-2r(t)d(\cos{\phi-wt})}]^3}(r-d\cos{(\phi-wt)})
$$

$$
\dot{p_r} = \frac{p_{\phi}^2 d^3}{md^3r^3} - \frac{GmM_Td^2}{d^2r^2} - \frac{Gmm_LM_T}{M_T[\sqrt{d^3(1+\frac{r^2(t)}{d^3}-\frac{2r(t)(\cos{\phi-wt}))}{d}}]^3}(r-d\cos{(\phi-wt)})
$$

$$
\dot{p_r} = \frac{\tilde{p_{\phi}}^2md}{\tilde{r}^3} - \frac{GmM_Td}{d^3\tilde{r}^2} - \frac{Gmm_LM_T}{M_T[\sqrt{(1+\frac{r^2(t)}{d^2}-\frac{2r(t)(\cos{\phi-wt}))}{d}}]^3}\frac{r-d\cos{(\phi-wt)}}{d^3}
$$
Tomando $\tilde{r′}=\sqrt{(1+\frac{r^2(t)}{d^2}-\frac{2r(t)(\cos{\phi-wt}))}{d})}$ y $\mu = \frac{m_L}{m_T}$


$$
\dot{p_r} = \frac{\tilde{p_{\phi}}^2md}{\tilde{r}^3} - \frac{GmM_Td}{d^3\tilde{r}^2} - \frac{Gm\mu M_T}{[\tilde{r′}]^3}\frac{\tilde{r}-\cos{(\phi-wt)}}{d^2}
$$
$$
\dot{p_r} = \frac{\tilde{p_{\phi}}^2md}{\tilde{r}^3} -\frac{GmM_Td}{d^3}(\frac{1}{\tilde{r}^2}  - \frac{\mu }{[\tilde{r′}]^3}\frac{\tilde{r}-\cos{(\phi-wt)}}{d^2})
$$

Tomando $\Delta = \frac{GM_T}{d^3}$

$$
\dot{p_r} = \frac{\tilde{p_{\phi}}^2md}{\tilde{r}^3} -\Delta md(\frac{1}{\tilde{r}^2}  - \frac{\mu }{[\tilde{r′}]^3}\frac{\tilde{r}-\cos{(\phi-wt)}}{d^2})
$$

Factorizando diviendo en ambos lados por $md$:
$$
\frac{\dot{p_r}}{md} = \frac{\tilde{p_{\phi}}^2}{\tilde{r}^3} -\Delta(\frac{1}{\tilde{r}^2}  - \frac{\mu }{[\tilde{r′}]^3}\frac{\tilde{r}-\cos{(\phi-wt)}}{d^2})
$$
$$
\dot{\tilde{p_r}} = \frac{\tilde{p_{\phi}}^2}{\tilde{r}^3} -\Delta(\frac{1}{\tilde{r}^2}  - \frac{\mu }{[\tilde{r′}]^3}\frac{\tilde{r}-\cos{(\phi-wt)}}{d^2})
$$
Para $\dot{p_\phi}= -G \frac{m m_L}{[r_L(r, \phi, t)]^3}(r(t)d[\sin{(\phi-wt)}])$

$$\dot{p_\phi}= -G \frac{m m_L}{[\sqrt{d^2+r^2(t)-2r(t)d(\cos{\phi-wt})}]^3}(r(t)d[\sin{(\phi-wt)}])$$

$$\dot{p_\phi}= -G \frac{m m_L M_T}{M_T[d\sqrt{(1+\frac{r^2(t)}{d^2}-\frac{2r(t)(\cos{\phi-wt}))}{d})}]^3}(r(t)d[\sin{(\phi-wt)}])$$

$$\dot{p_\phi}= -G \frac{m \mu M_T}{{d^3[\tilde{r′}}]^3}(r(t)d[\sin{(\phi-wt)}])$$
$$\dot{p_\phi}= - \frac{\Delta \mu md^2}{{[\tilde{r′}}]^3}\frac{(r(t)d[\sin{(\phi-wt)}])}{d^2}$$
$$\dot{p_\phi}= - \frac{\Delta \mu md^2}{{[\tilde{r′}}]^3}(\tilde{r(t)}[\sin{(\phi-wt)}])$$

Dividiendo en ambos lados $md$:

$$\frac{\dot{p_\phi}}{md^2}= - \frac{\Delta \mu}{{[\tilde{r′}}]^3}(\tilde{r(t)}[\sin{(\phi-wt)}])$$

$$\tilde{p_\phi}= - \frac{\Delta \mu}{{[\tilde{r′}}]^3}(\tilde{r(t)}[\sin{(\phi-wt)}])$$

Finalmente, para asignar los momentos canónicos iniciales muestre lo siguiente:


$$\tilde{p_r}=\frac{p_r}{md_L}= \frac{mv}{md_L}=\frac{1}{d_L}\frac{d r}{d t}$$

Dado que $r = \sqrt{x^2+y^2}$. Entonces:
$$\frac{1}{d_L}\frac{d \sqrt{x^2+y^2}}{d t}= \frac{1}{2d_L}\frac{1}{\sqrt{x^2+y^2}}\frac{d (x^2+y^2)}{d t}=\frac{1}{2d_L}\frac{1}{\sqrt{x^2+y^2}}(2x\dot{x} +2y\dot{y})$$

$$\tilde{p_r}=\frac{1}{d_L}\frac{1}{\sqrt{x^2+y^2}}(x\dot{x} +y\dot{y})=\frac{1}{d_L}\frac{1}{r}(x\dot{x} +y\dot{y})$$

Dado que $x=r\cos{\phi}$ y $y=r\sin{\phi}$. Y dado, $\vec{v}=(\dot{x}, \dot{y})=(v\cos{\theta},\sin{\theta})$. Por lo tanto, $\dot{x}=v\cos{\theta}$ y $\dot{y}=v\sin{\theta}$. Entonces se tiene la siguiente ecuación:

$$\tilde{p_r}=\frac{1}{d_L}\frac{1}{r}(r\cos{\phi}v\cos{\theta} +r\sin{\phi}\sin{\theta})=\frac{1}{d_L}(\cos{\phi}v\cos{\theta} +\sin{\phi}v\sin{\theta})$$

Factorizando $v$, da $\frac{v}{d_L} = \tilde{v}$

$$\tilde{p_r}=\tilde{v}(\cos{\phi}\cos{\theta} +\sin{\phi}v \sin{\theta})$$

Dado que $\cos{\phi}\cos{\theta}+\sin{\phi}\sin{\theta}=\cos{(\theta -\phi)}$

$$\tilde{p_r}=\tilde{v}\cos{(\theta -\phi)}$$

Considerando que en $t=t_0$:

$$\tilde{p_r}^0=\tilde{v_0}\cos{(\theta -\phi)}$$



$$\tilde{p_\phi}=\frac{p_\phi}{md_L^2}= \frac{mr^2\dot{\phi}}{md_L^2}=\frac{r^2}{d_L^2}\frac{d \phi}{d t}$$

$\phi$ está definido como $\phi =\arctan{\frac{y}{x}}$

$$\tilde{p_\phi}=\tilde{r}^2\frac{d (\arctan{\frac{y}{x}})}{d t}=\tilde{r}^2\frac{1}{1+\frac{y^2}{x^2}}\frac{d (\frac{y}{x})}{d t}=\tilde{r}^2\frac{1}{1+\frac{y^2}{x^2}}\frac{\dot{y}x-\dot{x}y}{x^2}$$

$$\tilde{p_\phi}=\tilde{r}^2\frac{1}{x^2+y^2}\dot{y}x-\dot{x}y$$

Dado que $r=\sqrt{x^2+y^2}$. Por lo tanto $r^2=x^2+y^2$:

$$\tilde{p_\phi}=\tilde{r}^2\frac{1}{r^2}\dot{y}x-\dot{x}y$$

Dado que $x=r\cos{\phi}$ y $y=r\sin{\phi}$. Y dado, $\vec{v}=(\dot{x}, \dot{y})=(v\cos{\theta},\sin{\theta})$. Por lo tanto, $\dot{x}=v\cos{\theta}$ y $\dot{y}=v\sin{\theta}$. Entonces se tiene la siguiente ecuación:

$$\tilde{p_\phi}=\tilde{r}^2\frac{1}{r^2}(v\sin{\theta}r\cos{\phi}-v\cos{\theta}r\sin{\phi})$$

Factorizando $r$ y $v$, además teniendo en cuenta que $\tilde{r}=\frac{r}{d_L}$

$$\tilde{p_\phi}=r^2\frac{vr}{d_L^2r^2}(\sin{\theta}\cos{\phi}-\cos{\theta}\sin{\phi})=\frac{vr}{d_L^2}(\sin{\theta}\cos{\phi}-\cos{\theta}\sin{\phi})=\tilde{r}\tilde{v}(\sin{\theta}\cos{\phi}-\cos{\theta}\sin{\phi})$$

Dado que $\sin{\theta}\cos{\phi}-\cos{\theta}\sin{\phi}=\sin{(\theta-\phi)}$

$$\tilde{p_\phi}=\tilde{r}\tilde{v}\sin{(\theta-\phi)}$$

```python
G = 6.67e-11
mT = 5.9736e24
rT = 6.3781e6
mL = 0.07349e24
rL = 1.7374e6
d = 3.844e8
omega = 2.6617e-6
```

```python
def System(Coord,t):
    
    r,phi,p_r,p_phi = Coord
    rm = np.sqrt(1+r**2 -2*r*np.cos(phi-omega*t))
    Del = G*mT/d**3
    mu = mL/mT
    dr_dt = p_r
    dphi_dt = p_phi/r**2
    dpr_dt = (p_phi)**2 / r**3 - Del*(1/r**2 + mu*(r-np.cos(phi-omega*t))/rm**3)
    dpphi_dt = (-Del*mu*r/(rm**3)) * np.sin(phi - omega*t)
    
    return np.array([dr_dt,dphi_dt, dpr_dt, dpphi_dt])
```

```python
def RK4(f,R,t):
    r0,phi0,pr0,pphi0=R
    h = t[1] - t[0]
    
    r = np.zeros_like(t)
    phi = np.zeros_like(t)
    pr = np.zeros_like(t)
    pphi = np.zeros_like(t)
    
    r[0] = r0
    phi[0] = phi0
    pr[0] = pr0
    pphi[0] = pphi0
    
    #K1 = np.zeros(4)
    #K2 = np.zeros(4)
    #K3 = np.zeros(4)
    #K4 = np.zeros(4)
    # Las pendientes instantaneas
    #R = np.zeros(4)
    for i in tqdm(range(1, len(t))):
        
        R = np.array([r[i-1],phi[i-1], pr[i-1],pphi[i-1]])
        K1 = f(R, t[i-1])
        
        R = np.array([r[i-1]+0.5*h*K1[0],phi[i-1]+0.5*h*K1[1], pr[i-1]+0.5*h*K1[2],pphi[i-1]+0.5*h*K1[3]])
        K2 = f(R, t[i-1]+0.5*h)
        
        R = np.array([r[i-1]+0.5*h*K2[0],phi[i-1]+0.5*h*K2[1], pr[i-1]+0.5*h*K2[2],pphi[i-1]+0.5*h*K2[3]])
        K3 = f(R, t[i-1]+0.5*h)
        
        R = np.array([r[i-1]+h*K3[0],phi[i-1]+h*K3[1], pr[i-1]+h*K3[2],pphi[i-1]+h*K3[3]])
        K4 = f(R, t[i-1]+h)
        
        r[i] = r[i-1] + h/6 * (K1[0]+2*K2[0]+2*K3[0]+K4[0])
        phi[i] = phi[i-1] + h/6 * (K1[1]+2*K2[1]+2*K3[1]+K4[1])
        pr[i] = pr[i-1] + h/6 * (K1[2]+2*K2[2]+2*K3[2]+K4[2])
        pphi[i] = pphi[i-1] + h/6 * (K1[3]+2*K2[3]+2*K3[3]+K4[3])

        
        #RK4 normal
        #Coeficientes
        #Valores iniciales
    return r, phi, pr, pphi
```

```python
G = 6.67e-11
mT = 5.9736e24
rT = 6.3781e6
mL = 0.07349e24
rL = 1.7374e6
d = 3.844e8
omega = 2.6617e-6
```

```python
t = np.arange(0,2.14e6,1)
```

```python
v0 = 11.1e3/d
phi_0 = 360*np.pi/180
ang = 30 *np.pi/180
r0 = 6.371e6/d

R0 = [r0, phi_0, v0*np.cos(ang-phi_0), v0*r0*np.sin(ang-phi_0)]
```

```python
sol2 = RK4(System,R0,t)
```

```python
r1 = sol2[0]
phi1 = sol2[1]
p_r1 = sol2[2]
p_phi1 = sol2[3]
xL1 = d*np.cos(omega*t)
yL1 = d*np.sin(omega*t)

xT1 = r1*np.cos(phi1)*d
yT1 = r1*np.sin(phi1)*d
```

```python
scale = 1000
```

```python
fig = plt.figure(figsize=(5,5))
ax = fig.add_subplot()


T = t[::scale]*(1/86400)

def init():
    
    ax.clear()
    ax.set_xlim(-4e8,4e8)
    ax.set_ylim(-4e8,4e8)

def Update(i):
    init()
        
    ax.scatter(xL1[::scale][i],yL1[::scale][i],marker=".", color='grey')
    ax.scatter(0,0,color="g")
    ax.set_title("t = {:.4f} días terrestres".format(T[i]))
    ax.plot(xL1[::scale][i],yL1[::scale][i], color='gray', linestyle="dashed")
    ax.plot(xT1[::scale][i],yT1[::scale][i], color='blue', linestyle="dashed")
    ax.scatter(xT1[::scale][i],yT1[::scale][i],marker=".",color="red")
        
Animation1 = anim.FuncAnimation(fig,Update,frames=len(t[::scale]),init_func=init)
```
