$a)$ Muestre que el error del método está descrito por:
$$\epsilon_{n+1} -(2+h^2 a′_{n})\epsilon_{n}+\epsilon_{n-1}=0. $$ $$a′_{n}=\frac{\partial a}{\partial x}  $$

Asumiendo que: $$x_{n} = \overline{x}_{n} +\epsilon_{n}$$

Por lo tanto $$x_{n+1} = \overline{x}_{n+1} +\epsilon_{n+1}$$
$$x_{n-1} = \overline{x}_{n-1} +\epsilon_{n-1}$$

Reemplazando en la ecuación de Verlet, $x_{n+1} = 2x_{n}-x_{n-1} + a_{n}(x_{n})h^2$, se tiene que:

$$\overline{x}_{n+1} +\epsilon_{n+1} = 2(\overline{x}_{n} +\epsilon_{n})-(\overline{x}_{n-1} +\epsilon_{n-1}) + a_{n}(\overline{x}_{n} +\epsilon_{n})h^2$$


Aproximando por Taylor $a_{n}(\overline{x}_{n} +\epsilon_{n}) \approx a_{n}(\overline{x}_{n}) +a′_{n}(\epsilon_{n})$, da como resultado:
$$\overline{x}_{n+1} +\epsilon_{n+1} = 2(\overline{x}_{n} +\epsilon_{n})-(\overline{x}_{n-1} +\epsilon_{n-1}) + a_{n}(\overline{x}_{n}) +a′_{n}(\epsilon_{n})h^2$$


Restandole a esta ecuación la ecuación de Verlet, da lo siguiente:
$$\epsilon_{n+1} = 2\epsilon_{n}-\epsilon_{n-1} + a′_{n}\epsilon_{n}h^2$$

Igualando a cero la ecuación, da como resultado:

$$\epsilon_{n+1} -(2+h^2 a′_{n})\epsilon_{n}+\epsilon_{n-1}=0 $$

$b)$ Para el caso de un oscilador armónico clásico muestre que: 
$$\epsilon_{n+1} -2(1-R)\epsilon_{n}+\epsilon_{n-1}=0 $$

donde $2R=h^2\omega^2$

La aceleración para un oscilador armónico esta dada por $a_{x}=-\omega^2 x$. Por lo tanto $$a′_{n}=\frac{\partial a}{\partial x}=-\omega^2   $$

Reemplazando $a′_{n}$ en la ecuación encontrada en $a)$, da como resultado:

$$\epsilon_{n+1} -(2-h^2 \omega^2)\epsilon_{n}+\epsilon_{n-1}=0 $$

Tomando $2R=h^2\omega^2$ y simplificando y factorizando se llega al resultado esperado:
$$\epsilon_{n+1} -(2-2R)\epsilon_{n}+\epsilon_{n-1}=0 $$

$$\epsilon_{n+1} -2(1-R)\epsilon_{n}+\epsilon_{n-1}=0 $$

$c)$ Usando la suposición de función potencia $\epsilon_{n}=\epsilon_{0}\lambda^n$, muestre que las raíces son:

$$\lambda_{\pm}=1-R \pm \sqrt{{R^2 - 2R}}$$

Si se supone que $\epsilon_{n}=\epsilon_{0}\lambda^n$, la ecuación queda de la siguiente manera:

$$\epsilon_{0}\lambda^{n+1} -(1-R)\epsilon_{0}\lambda^{n}+\epsilon_{0}\lambda^{n-1}=0 $$

Dividiendo entre $\lambda^{n-1}\epsilon_{0}$ la ecuación

$$\lambda^{2} -2(1-R)\lambda+1=0 $$
Utilizando la fórmula cuadrática, las soluciones para $\lambda$ son:

$$
\lambda_{\pm} = \frac{{-b \pm \sqrt{{b^2 - 4ac}}}}{{2a}}
$$

donde $a = 1$, $b = -2(1-R)$ y $c = 1$. Sustituyendo estos valores, obtenemos:

$$
\lambda_{\pm} = \frac{{2(1-R) \pm \sqrt{{(2(1-R))^2 - 4(1)(1)}}}}{{2(1)}}
$$

$$
\lambda_{\pm} = \frac{{2(1-R) \pm \sqrt{{4(1-R)^2 - 4}}}}{2}
$$

$$
\lambda_{\pm} = \frac{{2(1-R) \pm \sqrt{{4(1-2R+R^2) - 4}}}}{2}
$$

$$
\lambda_{\pm} = \frac{{2(1-R) \pm \sqrt{{4 - 8R + 4R^2 - 4}}}}{2}
$$

$$
\lambda_{\pm} = \frac{{2(1-R) \pm \sqrt{{4R^2 - 8R}}}}{2}
$$

$$
\lambda_{\pm} = \frac{{2(1-R) \pm 2\sqrt{{R^2 - 2R}}}}{2}
$$
Dando así el resultado esperado:
$$
\lambda_{\pm} = (1-R) \pm \sqrt{{R^2 - 2R}}
$$

$d)$ El valor de $\left| {\lambda_{\pm}}\right|  \leq 1$. define la estabilidad del algoritmo. Muestre que el paso de integración debe cumplir:

$$ h\leq \frac{2}{w}$$

Tomando el valor absoluto de la expresión para $\lambda_{\pm}$

$$\left| \lambda_{\pm} \right| = \left| (1-R) \pm \sqrt{R^2 - 2R} \right|$$

$$-1\leq(1-R) \pm \sqrt{R^2 - 2R}\leq 1$$

$$ R-2 \leq\pm\sqrt{R^2 - 2R}\leq R$$

Elevando al cuadrado la expresión

$$ (R-2)^2 \leq \left( \pm \sqrt{R^2 - 2R} \right)^2 \leq R^2 $$

Expandiendo el polinomio
$$ R^2 - 4R + 4 \leq R^2 - 2R \leq R^2 \quad  $$

$$ -2R \leq -4R +4 $$

Despejando $h$, da el resultado esperado:
$$ h\leq \frac{2}{w}$$

$e)$ Implemente una rutina sencilla del método de Verlet para el oscilador armónico y muestre en una animación el comportamiento en la región estable y en la región inestable. Use $x(0)=1$, $\dot{x}(0)=1$ y $\omega = \pi$ 
. Itere el tiempo suficiente para evidenciar el comportamiento de la estabilidad.

A continuación se puede apreciar un código que implementa el método de Verlet, más adelante se muestran gráficas que describen el comportamiento estable e inestable de éste algoritmo.

```python
def Harmonic(x, w):
    
    a= (-(w**2)*x )
    return a
```

```python
def GetVerlet(f,y0, v0,t,h, w):
    
    
    
    y = np.zeros_like(t)
    v = np.zeros_like(t)
    y[0] = y0
    v[0] = v0
    y[1] = y[0] + h*v[0] 
    v[1] = y[0] + h*f(y[1], w)
    
   # yc = y.copy()
    
    for i in range(2,len(y)):
        
        
        y[i] = 2*y[i-1] - y[i-2] + f(y[i-1],  w)*(h**2)
        v[i] = (y[i] - y[i-2])/(2*h)
    
    return y, v
```

Estable

```python
w = np.pi
h = 1/w
y0 = 1
v0 = 1
t=np.arange(0,20, 0.5)
```

```python
y,v=GetVerlet(Harmonic, y0, v0, t, h, w)
```

```python
fig = plt.figure(figsize=(10,5))
ax = fig.add_subplot(121)
ax1 = fig.add_subplot(122)

def init():
    
    ax.clear()
    ax.set_xlim(-3,3)
    ax.set_ylim(-3,3)
    
    ax1.clear()
    #ax1.set_xlim(-1,10)
    ax1.set_ylim(-2,2)
    
def Update(i):
    
    init()
    #ax.set_title(r't =  %.3f s' %(t1[i]))
    
    
    Position = y[i]
    Velocity = v[i]
        
    circle = plt.Circle( (0,Position), 0.3, color='r', fill=False )
    ax.add_patch(circle)
        
    ax.arrow(0,Position,0,Velocity,color='k',head_width=0.05,length_includes_head=True)
    ax1.plot( t[0:i], y[0:i], color='purple')
        
    #ax1.set_title(r'Total kinetic Energy: {:.3f}'.format(KE))
    #ax1.scatter(t1[:i], p.GetKineticEnergy(scale)[:i],color='k',marker='.')
        
Animation = anim.FuncAnimation(fig,Update,frames=len(t),init_func=init, repeat=False)
```
