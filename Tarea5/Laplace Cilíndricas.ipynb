$$
\nabla ^2 u(x,y) = \frac{\partial^2 u}{\partial x^2} +\frac{\partial^2 u}{\partial y^2}
$$


Toamando 
$
x = \rho \cos (\phi)
$
y
$ y = \rho \sin (\phi)$. Se eleva al cuadrado $x$ y $y$, y se suman:

$$
x^2 + y^2 = \rho^2 \cos^2(\phi) +\rho^2\sin^2(\phi) = \rho^2
$$

Por lo tanto:
$$
\rho = \sqrt{x^2 + y^2}
$$

Por otro lado, dividiendo $y$ entre $x$:

$$
\frac{y}{x} = \frac{\rho \sin (\phi)}{\rho \cos (\phi)}=\tan{\phi}
$$


Es decir que las derivadas de $\rho$ y $\phi$ con respecto a $x$ son:

$$
\phi = \arctan{\frac{y}{x}}
$$

$$
\frac{\partial \rho}{\partial x} = \frac{x}{\sqrt{x^2 + y^2}}
$$

$$
\frac{\partial \phi}{\partial x} = -\frac{y}{x^2 + y^2}
$$

$$
\frac{\partial u}{\partial x} = \frac{\partial u}{\partial \rho}\frac{\partial \rho}{\partial x} + \frac{\partial u}{\partial \phi}\frac{\partial \phi}{\partial x} 
\frac{\partial u}{\partial x} = \frac{\partial u}{\partial \rho}\frac{x}{\sqrt{x^2 + y^2}} + \frac{\partial u}{\partial \phi}(-\frac{y}{x^2 + y^2})
$$

Tomando la segunda derivada de $u$ con respecto a $x$:

$$
\frac{\partial}{\partial x}\frac{\partial u}{\partial x} =\frac{\partial}{\partial x}( \frac{\partial u}{\partial \rho}\frac{x}{\sqrt{x^2 + y^2}} + \frac{\partial u}{\partial \phi}(-\frac{y}{x^2 + y^2}))
$$


$$
\frac{\partial ^2 u}{\partial x^2} =\frac{\partial}{\partial x}(\frac{\partial u}{\partial \rho})\frac{x}{\sqrt{x^2 + y^2}} + \frac{\partial u}{\partial \rho}\frac{\partial}{\partial x}(\frac{x}{\sqrt{x^2 + y^2}})+ \frac{\partial}{\partial x}(\frac{\partial u}{\partial \phi})(-\frac{y}{x^2 + y^2}) +\frac{\partial u}{\partial \phi}\frac{\partial}{\partial x}(-\frac{y}{x^2 + y^2})
$$

$$
\frac{\partial ^2 u}{\partial x^2} =\frac{\partial}{\partial \rho}\frac{\partial \rho}{\partial x}\frac{\partial u}{\partial \rho}\frac{x}{\sqrt{x^2 + y^2}} +\frac{\partial u}{\partial \rho}(\frac{\sqrt{x^2 + y^2}-\frac{x^2}{\sqrt{x^2 + y^2}}}{x^2 + y^2}) -\frac{\partial}{\partial \phi}\frac{\partial \phi}{\partial x}\frac{\partial u}{\partial \phi}(\frac{y}{x^2 + y^2}) +\frac{\partial u}{\partial \phi}(\frac{2xy}{(x^2 + y^2)^2})

Sumando $\frac{\partial^2 u}{\partial x^2}$ y $\frac{\partial^2 u}{\partial y^2}$:
$$
\frac{\partial ^2 u}{\partial x^2} + \frac{\partial ^2 u}{\partial y^2} = \frac{\partial ^2 u}{\partial \rho^2} \cos^2(\phi) +\frac{\partial u}{\partial \rho}\frac{1-\cos^2(\phi)}{\rho}+\frac{\partial^2 u}{\partial \phi^2}\frac{\sin^2(\phi)}{\rho^2} +\frac{\partial u}{\partial \phi}\frac{2\cos(\phi)\sin(\phi)}{ \rho^2} + \frac{\partial ^2 u}{\partial \rho^2} \sin^2(\phi) +\frac{\partial u}{\partial \rho}\frac{1-\sin^2(\phi)}{\rho}+\frac{\partial^2 u}{\partial \phi^2}\frac{\sin^2(\phi)}{\rho^2} -\frac{\partial u}{\partial \phi}\frac{2\cos(\phi)\sin(\phi)}{ \rho^2}

Agrupando tèrminos:

$$

Simplificando:

$$
\frac{\partial ^2 u}{\partial x^2} + \frac{\partial ^2 u}{\partial y^2} = \frac{\partial ^2 u}{\partial \rho^2}  + \frac{\partial u}{\partial \rho}\frac{1}{\rho} + \frac{\partial^2 u}{\partial \phi^2}\frac{1}{\rho^2}
$$

$$
\nabla ^2u(\rho,\phi) = \frac{\partial ^2 u}{\partial \rho^2}  + \frac{\partial u}{\partial \rho}\frac{1}{\rho} + \frac{\partial^2 u}{\partial \phi^2}\frac{1}{\rho^2}
$$
